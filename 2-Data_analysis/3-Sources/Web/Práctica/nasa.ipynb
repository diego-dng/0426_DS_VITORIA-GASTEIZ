## Ejercicio 2: Extraer todos los asteroides que han pasado cerca de la tierra esta última semana. Tenéis que mostrar los datos en un DataFrame, y guardarlos en un CSV.

```python
URL_AST = "https://api.nasa.gov/neo/rest/v1/feed"
params = {
    "start_date": "2026-05-04",
    "end_date": "2026-05-08",
    "api_key": API_KEY
}

resp = requests.get(URL_AST, params=params)

if resp.status_code == 200:
    data = resp.json()
    json = json.dumps(data, indent= 5)
    print(json)
else:
    print("error:", resp.json().get("msg"))
```

```python
len(data["near_earth_objects"])
```

```python
lista_ast = []
for i in data["near_earth_objects"]:
    for j in data["near_earth_objects"][i]:
        #print("----------", j)
        info = {
            "id": j["id"],
            "nombre": j["name"],
            "magnitud": j["absolute_magnitude_h"],
            "km": j["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "metros": j["estimated_diameter"]["meters"]["estimated_diameter_min"],
            "km": j["estimated_diameter"]["feet"]["estimated_diameter_min"]
        }
        lista_ast.append(info)

print(lista_ast)
```

```python
import pandas as pd
df = pd.DataFrame(lista_ast)
df
```

```python
df.to_csv("datos_ast.csv")
```
