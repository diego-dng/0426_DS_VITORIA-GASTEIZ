### Step 4. Which continent drinks more beer on average?

```python
df.head()
```

```python
df["continent"].value_counts()
```

```python
df[df["continent"] == "NA"]
```

```python
df.groupby("continent")["beer_servings"].count().sum()
```

```python
df.groupby("continent")["beer_servings"].mean().idxmax()
```

```python
df[df["continent"].isnull()]
```

```python
df.loc[df["continent"].isnull(), "continent"] = "NA"
df.head(20)
```

```python
df["continent"].value_counts()
```

```python
df.groupby("continent")["beer_servings"].mean().idxmax()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
df.groupby("continent")["wine_servings"].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
df.groupby("continent").mean(numeric_only= True)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
df.groupby("continent").median(numeric_only= True)
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
df.groupby("continent")["spirit_servings"].aggregate(["min", "mean", "max"])
```
