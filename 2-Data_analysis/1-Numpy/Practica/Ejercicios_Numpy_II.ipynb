## Ejercicio 12
Encuentra los índices de los valores que no sean 0 del siguiente vector [1,2,0,0,4,0]. El output sería [0, 1, 4]. [Encuentra el método apropiado en la documentación.](https://numpy.org/doc/) Encapsúlalo en una función, con un argumento de entrada de un vector.

````python
x = np.array([1,2,0,0,4,0])

np.where(x != 0)[0]
````

````python
def indices(vector):
    return np.where(vector != 0)[0]
````

````python
resp = indices(x)
print(resp)
````

## Ejercicio 18
Crea una matriz aleatoria de 5x5 y normaliza sus valores. Para normalizar tienes que restar sus valores por la media(`np.mean`) y dividir el resultado por la desviación estándar(`np.std`).

````python
np.random.seed(11)
x = np.random.randint(1,20, (5,5))
print(x)
````

````python
np.mean(x)
````

````python
np.std(x)
````

````python
normalizar = (x -np.mean(x))/ np.std(x)
print(normalizar)
````

````python
np.std(normalizar)
````

## Ejercicio 19
1. Crea una secuencia de números que vayan del 0 al 10.
2. Crea un nuevo array, que parta del anterior, pero sustituya por el valor -1, todos los elementos cuyo valor se encuentre entre el 3 y el 8.
3. Crea un nuevo array que niegue todos los elementos entre el 3 y el 8

````python
x = np.arange(11)
print(x)
````

````python
(x >=3) & (x <= 8)
````

````python
x2 = x.copy()
x2[(x2 >=3) & (x2 <= 8)] = -1
print(x2)
````

````python
x3 = x.copy()
mas = (x3 >=3) & (x3 <= 8)
x3[mas] *= -1
print(x3)
````
